```python
import pandas as pd
```

```python
dfAttack = pd.read_csv(r"C:\Users\Neil Garcia\randomforest\dataCIC\attack1.csv")
```

```python
dfNormal = pd.read_csv(r"C:\Users\Neil Garcia\randomforest\dataCIC\normal1.csv")
```

```python
dfAttack["Label"] = 1
dfNormal["Label"] = 0
```

```python
data = [dfAttack, dfNormal]

for i in data:
    num_rows = len(i)
    print("Number of rows in the DataFrame:", num_rows)
```

```python
df = pd.concat([dfAttack, dfNormal], axis=0, ignore_index=True)
```

```python
# Shuffle the rows
df = df.sample(frac=1.0, random_state=42)
print(df)
```

```python
columns = df.columns
columns
```

```python
LFIFeatures = [
    'pkt_len_max', 'pkt_len_min', 'pkt_len_mean', 'pkt_len_std', 'pkt_len_var',
    'fwd_header_len', 'bwd_header_len', 'fwd_seg_size_min', 'fwd_act_data_pkts',
    'fwd_pkt_len_max', 'fwd_pkt_len_min', 'fwd_pkt_len_mean', 'fwd_pkt_len_std',
    'bwd_pkt_len_max', 'bwd_pkt_len_min', 'bwd_pkt_len_mean', 'bwd_pkt_len_std', 'Label'
]

df = df[LFIFeatures]
```

```python
# Calculate the number of rows in the DataFrame
num_rows = len(df)

# Calculate the number of rows for each half
half_num_rows = num_rows // 2

# Select the first half of the DataFrame
first_half_df = df.iloc[:half_num_rows]

# Select the second half of the DataFrame
second_half_df = df.iloc[half_num_rows:]

# Export the first half to a CSV file
first_half_df.to_csv('use1.csv', index=False)

# Export the second half to a CSV file
second_half_df.to_csv('use2.csv', index=False)
```
